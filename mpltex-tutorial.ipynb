# Tutorial for beautiful plots!

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def double_well(x):
    return 4 * (x**2 - 1)**2
```

```python
plt.figure(dpi=50, figsize=(10, 7))
xrange = np.linspace(-2, 2, num=100)
plt.plot(xrange, double_well(xrange), label='Ugly plot')
plt.xlabel(r'$\tau$')
plt.ylabel(r'$\Delta U(\tau)$')
plt.legend()
plt.show()
```

```python
import mpltex
```

```python
@mpltex.aps_decorator
def myplot():
    plt.figure(dpi=200)
    xrange = np.linspace(-2, 2, num=100)
    plt.plot(xrange, double_well(xrange), label='Beautiful plot')
    plt.xlabel(r'$\tau$')
    plt.ylabel(r'$\Delta U(\tau)$')
    plt.legend()
    plt.show()
    
myplot()
```
