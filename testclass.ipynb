```python
N = (2, 5, 10, 20, 50, 100)
t = list(())
for n in N:
    p = Polymer(n_beeds=n, mobility=0.000005)
    p.create_chain(axis=0)
    t1 = time()
    p.simulate(steps=3000)
    t2 = time()
    t.append(t2-t1)
    
plt.figure()
plt.plot(N, t)
plt.show()
```
